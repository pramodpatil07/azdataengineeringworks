# azpyworks example

This notebook demonstrates how to add `azpyworks` to `sys.path`, load shared configuration, and use notebook utilities in Databricks.

```python
import sys
from pathlib import Path

# Update this path to your Databricks repo location.
repo_root = '/Workspace/Repos/<your-repo-path>'
package_root = str(Path(repo_root) / 'azpyworks')
if package_root not in sys.path:
    sys.path.insert(0, package_root)

from azpyworks import load_config, display_dataframe

config = load_config(str(Path(package_root) / 'configs' / 'sample_config.json'))
print('Loaded config:')
print(config)
```

## Display example

If you have a pandas DataFrame, `display_dataframe` will print it nicely. If you have a Spark DataFrame, it will use `.show()`.

```python
try:
    import pandas as pd
    example_df = pd.DataFrame({
        'id': [1, 2],
        'name': ['Alice', 'Bob']
    })
    display_dataframe(example_df)
except ImportError:
    print('Pandas not available; if using Spark, use a Spark DataFrame instead.')
```
